```python
data = pd.read_csv("/content/SuperMarket Analysis.csv")
data
```

```python
data.isnull().sum()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=data, x='Quantity', y='Sales', hue='Product line', palette='Set1')
plt.title("Quantity vs Sales")
plt.xlabel("Quantity")
plt.ylabel("Sales")
plt.tight_layout()
plt.show()
```
